```python
import pandas as pd

cols = [
    "age", "sex", "cp", "trestbps", "chol", "fbs",
    "restecg", "thalach", "exang", "oldpeak",
    "slope", "ca", "thal", "target"
]

df = pd.read_csv("/Users/victorchen/Desktop/Cardio-Risk-Prediction/data/heart.csv", header=None, names=cols)


# Replace '?' with NaN
df.replace("?", pd.NA, inplace=True)

# Drop rows with missing values (simple approach)
df.dropna(inplace=True)

# Convert numeric columns
for col in df.columns:
    df[col] = pd.to_numeric(df[col])

# Convert target to binary (0 vs 1)
df["target"] = (df["target"] > 0).astype(int)

df.to_csv("../data/heart_clean.csv", index=False)

print("Saved cleaned dataset:", df.shape)
```
